```python
import pandas as pd
```

```python
#read the data
df = pd.read_csv('birdstrikes.csv')
df["Airport Name"].unique()
df
```

```python
#read the data
df1 = pd.read_csv('airport.csv')

#negate logitude values to make them negative
#already negative values should not be changed
df1['longitude'] = df1['longitude'].apply(lambda x: -x if x > 0 else x)
df1
```

```python
lats = []
longs = []
for i in df["Airport Name"]:
    for j in df1["airport"]:
        if i == j:
            lats.append(df1[df1["airport"] == i]["latitude"].values[0])
            longs.append(df1[df1["airport"] == i]["longitude"].values[0])
#add the new columns to the dataframe
df["latitude"] = lats
df["longitude"] = longs
df
```

```python
#take columns that are needed (AIRPORT NAME, STATE, LATITUDE, LONGITUDE)
df = df[["Airport Name", "Origin State", "latitude", "longitude"]]
#add a frequency column
df["frequency"] = 1
#count friquency of each airport
df = df.groupby(["Airport Name", "Origin State", "latitude", "longitude"]).count().reset_index()
df
```

```python
df.to_csv("dashboard/data/frequency.csv", index=False)
```
